```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
print(numerical_columns)
# Fill NaN values in numerical columns with the median
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].median())
df.isnull().sum()
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

unique_values_freq = {}

for col in categorical_columns:
    unique_values_freq[col] = df[col].value_counts()

# Display the unique values and their frequencies
for col, freq in unique_values_freq.items():
    print(f"Unique values and frequencies for column '{col}':\n{freq}\n")
```

```python
for col in categorical_columns:
    value_counts = df[col].value_counts()
    to_drop = value_counts[value_counts < 10].index
    df[col] = df[col].replace(to_drop, pd.NA)  # Replace infrequent categories with NaN

# Drop rows with NaN in any categorical column after replacement
df = df.dropna(subset=categorical_columns)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create the OneHotEncoder instance
encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' to avoid dummy variable trap

# Fit and transform the categorical columns
encoded_columns = encoder.fit_transform(df[categorical_columns])

# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(categorical_columns))

# Concatenate the original DataFrame (without categorical columns) with the encoded columns
df_enc = pd.concat([df.drop(categorical_columns, axis=1).reset_index(drop=True), encoded_df.reset_index(drop=True)], axis=1)
df_enc.head()
```

```python
import numpy as np
import pandas as pd

def assign_numeric_bins(df, column_bins):
    for column, num_bins in column_bins.items():
        min_val = df[column].min()
        max_val = df[column].max()
        
        # Create bins using linspace
        bins = np.linspace(min_val, max_val, num_bins + 1)
        
        # Create labels for each bin automatically
        labels = range(1, num_bins + 1)
        
        # Assign the numeric values based on the bins directly to the column
        df[column] = pd.cut(df[column], bins=bins, labels=labels, right=False)
        
        # Fill NaN values with 0 or another default integer value, then convert to int
        df[column] = df[column].cat.add_categories(0).fillna(0).astype(int)

# Dictionary of columns with their corresponding number of bins
columns_with_bins = {
    'Sleep_Hours': 3,
    'Previous_Scores': 3,
    'Attendance': 4,
    'Hours_Studied': 6
}

# Call the function with your DataFrame
assign_numeric_bins(df_enc, columns_with_bins)
```

```python
X = df_enc.drop(columns = ["Exam_Score"])
y = df_enc["Exam_Score"]
print("Block Executed")
```

```python
from xgboost import DMatrix, cv, XGBRegressor
from skopt import BayesSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb

# Split data into train/validation/test sets
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=9)

# Convert to DMatrix format for XGBoost
dtrain = DMatrix(data=X_train, label=y_train)
dval = DMatrix(data=X_val, label=y_val)
dtest = DMatrix(data=X_test, label=y_test)

# Enhanced hyperparameter space for tuning with BayesSearchCV
xgboost_params = {
    'n_estimators': (300, 1000),         # Increase range for more tuning flexibility
    'max_depth': (4, 12),                # More depth for complex patterns
    'learning_rate': (0.005, 0.3, 'log-uniform'),
    'subsample': (0.6, 1.0, 'uniform'),  # Prevent overfitting by sampling
    'colsample_bytree': (0.5, 1.0, 'uniform'),
    'min_child_weight': (1, 10),         # Minimum sum of instance weight (for regularization)
    'gamma': (0, 5),                     # Minimum loss reduction
    'lambda': (0.1, 10, 'log-uniform'),  # L2 regularization
    'alpha': (0.1, 10, 'log-uniform')    # L1 regularization
}

# Initialize XGBoost model with GPU settings
xgb_model = XGBRegressor(
    tree_method='hist',  # GPU-accelerated histogram algorithm
    random_state=9       # Reproducibility
)

# Use BayesSearchCV for hyperparameter tuning
xgb_opt = BayesSearchCV(
    estimator=xgb_model,
    search_spaces=xgboost_params,
    n_iter=30,                  # Increase iterations for better optimization
    cv=3,                       # 3-fold CV
    n_jobs=-1,                  # Use all cores
    scoring='neg_mean_squared_error',
    random_state=9
)

# Fit the model with BayesSearchCV
xgb_opt.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=0)

# Best parameters after tuning
print("Best hyperparameters:", xgb_opt.best_params_)

# Retrieve the best hyperparameters from the tuned model
best_params = xgb_opt.best_params_
best_params.update({
    'objective': 'reg:squarederror',  # Ensure the objective matches
    'eval_metric': 'rmse'            # Metric for evaluation
})

# Perform cross-validation using the best hyperparameters
cv_results = cv(
    params=best_params,
    dtrain=dtrain,
    num_boost_round=500,  # Fixed number of boosting rounds
    nfold=3,
    metrics='rmse',
    as_pandas=True,
    seed=9
)

# Display cross-validation results
print("Cross-validation RMSE results:\n", cv_results[['train-rmse-mean', 'test-rmse-mean']])

# Final evaluation on the test set
best_model = xgb_opt.best_estimator_
y_test_pred = best_model.predict(X_test)
test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)
test_r2 = r2_score(y_test, y_test_pred)

print(f"Test RMSE: {test_rmse}")
print(f"Test R^2 Score: {test_r2}")
```

```python
import matplotlib.pyplot as plt
import shap

# Learning Curve
def plot_learning_curve(cv_results):
    # Extracting train and test RMSE values
    train_rmse = cv_results['train-rmse-mean']
    val_rmse = cv_results['test-rmse-mean']
    
    plt.figure(figsize=(10, 6))
    plt.plot(train_rmse, label='Training RMSE')
    plt.plot(val_rmse, label='Validation RMSE')
    plt.xlabel('Boosting Round')
    plt.ylabel('RMSE')
    plt.title('Learning Curve: Training vs. Validation RMSE')
    plt.legend()
    plt.show()

# Call the function to plot learning curve
plot_learning_curve(cv_results)

# Feature Importance with SHAP
def plot_shap_feature_importance(model, X_train):
    # Initialize SHAP explainer using the model and data
    explainer = shap.Explainer(model, X_train)
    shap_values = explainer(X_train)

    # Summary plot of SHAP values (feature importance)
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values, X_train, plot_type="bar", show=True)
    shap.summary_plot(shap_values, X_train, show=True)

# Plot SHAP feature importance
plot_shap_feature_importance(xgb_opt.best_estimator_, X_train)
```
